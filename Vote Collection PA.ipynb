```python
try:
    wb = opx.load_workbook('PA votes.xlsx')
except:
    wb = opx.Workbook()

ws = wb.active

chmaber = ['S','H']

for c in chmaber:
    if c == 'S':
        print('Starting Senate')
    else:
        print('Starting House')
        
    iter_up = 0    
    exit = 1
    
    while exit == 1:
        
        iter_up += 1
                
        webad = f'https://www.legis.state.pa.us/CFDOCS/Legis/RC/Public/rc_view_action2.cfm?sess_yr=20{y}&sess_ind=0&rc_body={c}&rc_nbr={iter_up}'
        res = req.get(webad)
        tree = html.fromstring(res.content)
        if c == 'S':
            bill = tree.xpath('//*[@id="PageRollCallView"]/section/div/div[3]/div[2]/div[2]/a/text()')
        else:
            bill = tree.xpath('//*[@id="PageRollCallView"]/section/div/div[3]/div[2]/div[3]/a/text()')
        
        #Exit out once no roll calls are presented
        try:
            get_out = tree.xpath('//*[@id="PageRollCallView"]/section/div/h3/text()')
            if get_out[0].find("No Roll Calls found") > 0:
                exit = 0
                print(f'Exiting...ended on {iter_up}')
        
        except:
            pass
            
        
        if (iter_up % 25) == 0:
            print(f'~~~~~Passed {iter_up}...')
        
        #print(bill)

        if bill in acu_bills:
            soup = bsoup(res.content)
            yea_nay = soup.select('#RCVotesSum')
            print(yea_nay[0].text)

            votes = soup.select('.Column-ThreeFourth div')
            
            try:
                ws = wb[f'{c} {bill[0]}']
            except:
                ws = wb.create_sheet(f'{c} {bill[0]}')
            
            ws_max_row = ws.max_row + 2
            
            
            yea_nay = yea_nay[0].text
            yea_nay = yea_nay.replace("YEAS", "YEAS ")
            yea_nay = yea_nay.replace("NAYS", "NAYS ")
            yea_nay = yea_nay.replace("\n", " ")            

            ws[f'A{ws_max_row}'] = yea_nay
            
            ws_max_row = ws.max_row + 1
            
            for item in votes:
                if len(item) > 100:
                    pass
                else:
                    single_vote = item.text
                    single_vote = single_vote.replace('\n', '')
                    single_vote = single_vote.replace('\r', '')
                    single_vote = single_vote.replace('\t', '')


                    if single_vote.find('Y ') == 0:
                        vote_cast = 'Y'
                    elif single_vote.find('N ') == 0:
                        vote_cast = 'N'
                    else:
                        vote_cast = 'X'

                    single_vote = single_vote[2:]

                    ws[f'A{ws_max_row}'] = single_vote
                    ws[f'B{ws_max_row}'] = vote_cast

                    ws_max_row += 1

            print(f'{bill[0]} in')
        wb.save('PA votes.xlsx')
        
        
wb.save('PA votes.xlsx')
wb.close()
    
print('Done')
```
